```python
import pandas as pd
import glob
```

```python
df = pd.DataFrame()
for filename in glob.iglob('./mol_csv/mol_*.csv'):
    df = df.append(pd.read_csv(filename, index_col=0))
```

"Склеиваем" все .csv файлы в один.

```python
df
```

```python
data = df.copy()
data = data[data.f_value.notnull()] 
data = data[data.h_value.notnull()]
data = data[data.f_units.notnull()]
data = data[data.h_units.notnull()]
data = data[data.f_relation.notnull()]
data = data[data.h_relation.notnull()]

data = data[data.f_value != 0]
data = data[data.h_value != 0]

data = data[data.f_relation == '=']
data = data[data.h_relation == '=']

data = data[data.target_type == 'SINGLE PROTEIN']

data = data[data.h_units == data.f_units]


data = data.reset_index(drop=True)

data.result = data.f_value / data.h_value
```

Берем только корректные данные: где указано значение, где значение не 0. Также оставляем только те строки, где значения констант со знаком =, это сделано для упрощения и наделения смыслом подсчета result, т.к. не совсем очевидно, что делать, если для молекулы с водородом константа > x, а для молекулы с фтором < y. Оставляем только строки, где совпадают единицы измерения (units).

```python
data[['f_value', 'h_value', 'result']]
```

```python
cols = list(data.columns)
cols = [col for col in cols if not col.endswith('_id')]
data = data[cols]
```

```python
data.to_csv('clear_single_protein_decimal_comma.csv', decimal=",", sep=";")
```
